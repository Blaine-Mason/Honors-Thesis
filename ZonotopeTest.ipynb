```python
import numpy as np
from matplotlib import pyplot as plt
import itertools
from scipy.spatial import ConvexHull, convex_hull_plot_2d
```

```python
def generate_Z(A,n,axs,i,j):
    l = list(itertools.product([0, 1], repeat=n))
    x = np.array(l)
    z = A@x.T
    #print(z)
    #plt.plot(z[0,:], z[1,:], "o")
    #taken from docs
    hull = ConvexHull(z.T)
    for simplex in hull.simplices:
        axs[i, j].plot(z.T[simplex, 0], z.T[simplex, 1], 'r--')
```

```python
d = 2
n = 3
#taken from docs
fig, axs = plt.subplots(2, 2, gridspec_kw={'hspace': .4, 'wspace': .4})
i = 0
j = 0
A = np.random.uniform(0, 1, (2,n))
generate_Z(A,n,axs,i,j)
axs[0, 0].set_title('Uniform(0,1)')

j += 1
A = np.random.standard_cauchy((2,n))
generate_Z(A,n,axs,i,j)
axs[0, 1].set_title('Cauchy(0,1)')

i += 1
j = 0
A = np.random.normal(0,1,(2,n))
generate_Z(A,n,axs,i,j)
axs[1, 0].set_title('Normal(0,1)')

j += 1
A = np.random.beta(1,3,(2,n))
generate_Z(A,n,axs,i,j)
axs[1, 1].set_title('Beta(1,3)')

#A = np.random.uniform(0, 1, (2,n))
#A = np.array([[0, 3, 4], [4, 3, 1]])
#generate_Z(A,n,axs,i,j)
#for i in range(0, n):
    #generator = A[:, i].T
    #vector = np.array([[0,0], generator])
    #plt.plot(vector[:, 0], vector[:,1], '-')

plt.show()
```
